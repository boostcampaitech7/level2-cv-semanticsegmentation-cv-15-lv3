```python
import numpy as np
import pickle

# Define dataset with pairs (e.g., case_0 with case_1, case_2 with case_3, ...)
pairs = [[f'case_{2*i:04d}', f'case_{2*i+1:04d}'] for i in range(400)]  # Adjust for your total pairs
print(f"Total pairs: {len(pairs)}")

# Flatten to get all cases
all_cases = [case for pair in pairs for case in pair]

# Shuffle pairs (not individual cases) for random splitting
np.random.seed(42)  # For reproducibility
np.random.shuffle(pairs)

# Define splits (e.g., 5-fold cross-validation)
num_folds = 5
splits = []
fold_size = len(pairs) // num_folds

for i in range(num_folds):
    val_pairs = pairs[i * fold_size: (i + 1) * fold_size]
    train_pairs = [pair for pair in pairs if pair not in val_pairs]
    
    # Flatten pairs back to case lists
    val_cases = [case for pair in val_pairs for case in pair]
    train_cases = [case for pair in train_pairs for case in pair]
    
    # Convert to numpy arrays for the desired format
    splits.append({
        'train': np.array(train_cases),
        'val': np.array(val_cases)
    })

# Save the splits to `splits_final.pkl`
output_path = '/data/ephemeral/home/nnUNet_preprocessed/Task001_HandBone/splits_final.pkl'
with open(output_path, 'wb') as f:
    pickle.dump(splits, f)

print(f"New splits_final.pkl file created at: {output_path}")

# Print split summary in the specified format
for i, split in enumerate(splits):
    print(f"Fold {i} - Train: {split['train']}, Val: {split['val']}")
```
